```python
import numpy as np
import pandas as pd

train_df = pd.read_csv("../../data/viwikifc/train_final_pro.csv")
dev_df = pd.read_csv("../../data/viwikifc/dev_final_pro.csv")
test_df = pd.read_csv("../../data/viwikifc/test_final_pro.csv")

df = pd.concat([train_df, dev_df, test_df], ignore_index=True)
df.head()
```

```python
main_df = df[["claim", "context", "gold_label"]]
main_df.info()
```

```python
main_df.gold_label.value_counts()
```

```python
main_df[main_df['gold_label'] == "Supports"].to_csv('../../data/viwikifc/viwikifc_supported.csv', index=False)
```

```python
main_df[main_df["gold_label"].isin(["Supports", "Refutes"])].to_csv('../../data/viwikifc/viwikifc_relevant_pairs.csv', index=False)
main_df[main_df["gold_label"].isin(["Supports", "Refutes"])].shape
```
