```python
dataset_path = "/content/drive/MyDrive/KLTN_Project/Datasets"
dp_run_file = f"{dataset_path}/deeperimpact_vifc/run_file.txt"
bm25_run_file = f"{dataset_path}/bm25_run_file.txt"
test_qrels = f"{dataset_path}/test_qrels.tsv"
```

```python
%cd /content/fact-checking-data-framework-vn/

for i in range(1, 10):
    current_out_temp = f"/content/run_file_{i}.txt"
    print(f"\n--- Running with alpha=0.{i} and output to {current_out_temp} ---")
    get_ipython().system('python -m src.scripts.score_hybrid \
      --bm25_run {bm25_run_file} \
      --deep_impact_run {dp_run_file} \
      --output_file {current_out_temp} \
      --alpha 0.{i}')
```

```python
%cd /content/improving-learned-index

for i in range(1, 10):
    current_out_temp = f"/content/run_file_{i}.txt"
    print(f"\n--- Evaluating with alpha=0.{i} ---")
    get_ipython().system('python -m src.deep_impact.evaluate \
      --run_file_path {current_out_temp} \
      --qrels_path {test_qrels}')

# This cell is now incorporated into the previous cell for the loop. It will be removed or commented out.
# If you need to run evaluation separately for a specific file, you can uncomment and modify this.
# !python -m src.deep_impact.evaluate \
#   --run_file_path {out_temp} \
#   --qrels_path {test_qrels}
```
