```python
dataset_path = "/content/drive/MyDrive/KLTN_Project/Datasets"
output_expansions = f"{dataset_path}/deeperimpact/expansions.csv"
```

```python
docs = f"{dataset_path}/unique_doc_mapping.csv"
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
expansion_df = pd.read_csv(output_expansions)
expansion_df.head()
```

```python
docs_df = pd.read_csv(docs)
docs_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the length of expansion terms
expansion_df['expansion_length'] = expansion_df['expansion_terms'].apply(lambda x: len(str(x).split()))

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(expansion_df['expansion_length'], bins=50, kde=True)
plt.title('Distribution of Expansion Token Lengths')
plt.xlabel('Number of Tokens')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
print("Max token: ", expansion_df['expansion_length'].max())
print("Min token: ", expansion_df['expansion_length'].min())
print("Mean token: ", expansion_df['expansion_length'].mean())
```
