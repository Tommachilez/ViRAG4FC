# Load ViFactCheck

```python
import numpy as np
import pandas as pd
```

```python
vifactcheck_df = pd.read_csv('../../data/vifactcheck/vifactcheck_extract_data.csv')
vifactcheck_df.head()
```

```python
# Save label 0 and 1 to a separate csv file
vifactcheck_df[vifactcheck_df.label.isin([0, 1])].to_csv('../../data/vifactcheck/vifactcheck_relevant_pairs.csv', index=False)
```

```python
vifactcheck_df[vifactcheck_df.label.isin([0, 1])].shape
```

```python
import json

vifactcheck_unique = vifactcheck_df['context'].unique()

# Save unique contexts to a JSON file with idx having prefix 'c_'
with open('../../data/vifactcheck_unique_contexts.json', 'w') as f:
    json.dump({f'c_{i}': context for i, context in enumerate(vifactcheck_unique)}, f, indent=4)
```
