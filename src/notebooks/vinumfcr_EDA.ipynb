```python
import json
import numpy as np
import pandas as pd

def load_jsonl(file_path):
    """Load a JSONL file and return a list of dictionaries."""
    with open(file_path, 'r', encoding="utf8") as f:
        data = [json.loads(line) for line in f]
    return data

train_data = load_jsonl("../../data/vinumfcr/train_evidence.jsonl")
dev_data = load_jsonl("../../data/vinumfcr/dev_evidence.jsonl")
test_data = load_jsonl("../../data/vinumfcr/test_evidence.jsonl")
data = train_data + dev_data + test_data
```

```python
df = pd.DataFrame(
    data=data,
)
```

```python
main_df = df[["claim", "context", "label"]]
```

```python
main_df.info()
```

```python
main_df[main_df.label=="Supported"].to_csv('../../data/vinumfcr/vinumfcr_supported.csv', index=False)
```

```python
main_df[main_df.label.isin(["Supported", "Refuted"])].to_csv('../../data/vinumfcr/vinumfcr_relevant_pairs.csv', index=False)
```
